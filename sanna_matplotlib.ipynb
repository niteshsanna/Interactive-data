1. Import the libraries

```python
# Importing the libaries
import pandas as pd
import matplotlib.pyplot as plt
```

### Data Analyst – Parallel Sets Plot
Description: A parallel sets plot displays multivariate data by showing data points across several axes, enabling you to see correlations between different variables. A data analyst could use this to analyze customer profiles, looking at correlations between age, income, spending, and satisfaction levels to detect patterns among different customer groups.




3. Loading and Viewing the dataset

```python
# Load the dataset
community_data = pd.read_csv('cces_example_data.csv')
```

```python
community_data.head()
```


4. Creating an empty figure featuring two axes organized in a matrix layout, using the methods and syntax covered in class. Naming the first axis as "Age Histogram" and the second as "Political Identity."

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Set titles for the axes
axs[0].set_title('Age Histogram')
axs[1].set_title('Political Identity')

# Show the empty plot
plt.tight_layout()
plt.show()
```


5. Creating a figure with one axis that displays a bar plot showing the frequency of responses for the Evaluation variable.

```python
# Summarize and plot the frequencies of the Evaluation variable
evaluation_counts = community_data['Evaluation'].value_counts()

fig, ax = plt.subplots(figsize=(8, 5))
evaluation_counts.plot(kind='bar', ax=ax)
ax.set_title('Frequencies of Evaluation Responses')
ax.set_xlabel('Evaluation')
ax.set_ylabel('Frequency')

plt.show()
```


6. Creating a figure with a single axis that displays a scatter plot illustrating the relationship between the ConMediaUse and ImmiPhobia variables.

```python
# Scatter plot for ConMediaUse and ImmiPhobia
fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(community_data['ConMediaUse'], community_data['ImmiPhobia'])
ax.set_title('ConMediaUse vs. ImmiPhobia')
ax.set_xlabel('ConMediaUse')
ax.set_ylabel('ImmiPhobia')

plt.show()
```


7. Improvements for Scatterplot

- Add Labels for Key Points: Labeling important or key data points (e.g., highest or lowest values) can help emphasize particular insights. This could be the highest revenue point or a significant observation in the data.
- Utilize size or color mapping: Adding extra dimensions, such as color or size, can provide more depth to the information presented. For instance, using color to represent a third variable, like age or gender, can help uncover patterns among multiple variables.
